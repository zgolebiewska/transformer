```python
from transformers import pipeline, AutoModelForTokenClassification, AutoTokenizer
import pandas as pd
from tqdm.notebook import tqdm
```

```python
train_data = pd.read_csv('./train/train.tsv', sep='\t', header=None)
X_test = pd.read_csv('./dev-0/in.tsv', sep='\t', header=None)
Y_test = pd.read_csv('./dev-0/expected.tsv', sep='\t', header=None)
```

```python
X_train = train_data[train_data.columns[1]]
Y_train = train_data[train_data.columns[0]]
X_test = X_test[X_test.columns[0]]
Y_test = Y_test[Y_test.columns[0]]
```

```python
X_train = [text.split() for text in X_train]
X_test = [text.split() for text in X_test]
```

```python
model = AutoModelForTokenClassification.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-cased")
```

```python
recognizer = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
def process_tokens(tokens):
    processed_tokens = []
    current_word = ""
    current_tag = None
    
    for token in tokens:
        if token['word'].startswith("##"):
            current_word += token['word'][2:]
        else:
            if current_word:
                processed_tokens.append((current_word, current_tag))
            current_word = token['word']
            current_tag = token['entity']
    
    if current_word:
        processed_tokens.append((current_word, current_tag))
    
    return processed_tokens
```

```python
full_tags = []
for idx, sentence in tqdm(enumerate(X_test)):
    predicted_tags = []
    corrected_tags = []
    ner_results = recognizer(" ".join(sentence))
    processed_tokens = process_tokens(ner_results)
    
    tag_idx = 0
    for _ in Y_test[idx].split():
        if tag_idx < len(processed_tokens):
            predicted_tags.append(processed_tokens[tag_idx][1])
            tag_idx += 1
        else:
            predicted_tags.append("O")
    
    predicted_tags = ["B-" + tag[2:] if tag and tag.startswith("I-") else tag for tag in predicted_tags]
    
    for i, tag in enumerate(predicted_tags):
        if i > 0 and tag == predicted_tags[i-1]:
            corrected_tags.append(tag.replace("B-", "I-"))
        else:
            corrected_tags.append(tag)
    
    full_tags.append(corrected_tags)
```

```python
with open("out.tsv", 'w') as file:
    for tags in full_tags:
        # Filter out None values and convert non-None elements to strings
        tags = [tag if tag is not None else "" for tag in tags]
        file.write(" ".join(tags) + "\n")

print(f"Processed {len(full_tags)} sentences.")
```
